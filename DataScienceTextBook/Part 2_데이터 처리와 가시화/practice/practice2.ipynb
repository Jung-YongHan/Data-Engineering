matplotlib.pyplot - 다수의 그래프 작성 도구를 제공한다.
- 웹 사이트: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html
- 이 웹 사이트에는 acorr, axes, bar, figure, gcf, hist, imread, plot, savefig, scatter, tick_params, xlim, ylim 등의 설명이 있다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

fig = plt.subplots(figsize=(10,3)) # size[inch, inch]
plt.plot(x, y2)
plt.plot(x, y3)

plt.grid()
plt.xlabel('x')
plt.xlabel('y')
plt.title('Example')


```

```python
# 하나의 그래프에 네 개의 좌표축을 그림
# 방법 1
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(6,4))

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
```
